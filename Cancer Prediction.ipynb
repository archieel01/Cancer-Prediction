```python
import pandas as pd
```

```python
cancer=pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
cancer.info()
```

```python
cancer.head(10)
```

```python
cancer.describe()
```

```python
cancer.columns
y=cancer['diagnosis']
X=cancer.drop(['id', 'diagnosis', 'Unnamed: 32'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=5000)
```

```python
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize=(8, 6))
disp = ConfusionMatrixDisplay(conf_matrix, display_labels=model.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```

```python
feature_names = X.columns
coef = model.coef_.flatten()
feature_importance = pd.Series(coef, index=feature_names.sort_values(ascending =False))
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importance.values, y=feature_importance.index, palette='viridis')
plt.title('Feature Importance')
plt.show()
```

```python
max_importance = feature_importance.values.max()
imp = feature_importance[feature_importance==max_importance]
print(imp)
```
